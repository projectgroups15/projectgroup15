```python
import sqlite3 
import csv 
```

```python
con = sqlite3.connect('student_lifestyle_dataset.db')
cursor = con.cursor()
```

```python
create_table = "CREATE TABLE Students_lifestyle (Student_ID FLOAT, Study_Hours_Per_Day FLOAT, Extracurricular_Hours_Per_Day FLOAT, Sleep_Hours_Per_Day FLOAT, Social_Hours_Per_Day FLOAT, Physical_Activity_Hours_Per_Day FLOAT, GPA FLOAT, Stress_Level VARCHAR(50));"
cursor.execute(create_table)
```

```python
with open("student_lifestyle_dataset.csv") as sl:
    data = csv.reader(sl)
    next(data)
    for row in data:
        cursor.execute("INSERT INTO Students_lifestyle VALUES (?,?,?,?,?,?,?,?)", (float(row[0]), float(row[1]), float(row[2]), float(row[3]), float(row[4]), float(row[5]), float(row[6]), row[7]))
```

```python
cursor.execute("SELECT * FROM Students_lifestyle").fetchall()
```

```python
con.commit()
```
